# User‑based Collaborative Filtering на MovieLens 1M

## 1. Импорт библиотек и загрузка данных

```python
import sys
import os
sys.path.append(os.path.abspath("."))

from src.data_loader import DataLoader
from src.metrics import Metrics
from src.utils import train_test_split_by_user

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances

# Загрузка данных
loader = DataLoader(path='./data')
ratings, movies, users = loader.load_all()

print("Рейтингов:", len(ratings))
print("Пользователей:", ratings['userId'].nunique())
print("Фильмов:", ratings['movieId'].nunique())
```

## 2. Разделение на train / tes

```python
train, test = train_test_split_by_user(ratings, test_size=0.2, seed=42)
print(f"Train: {len(train)} оценок, Test: {len(test)} оценок")
```

## 3. Построение матрицы пользователи → фильмы и центрирование

```python
# Создаём матрицу user–item (оценки, NaN где нет)
user_item_matrix = train.pivot(index='userId', columns='movieId', values='rating')
print("Размер матрицы:", user_item_matrix.shape)

# Вычисляем среднюю оценку каждого пользователя
user_means = user_item_matrix.mean(axis=1)
# Центрируем (вычитаем среднее) – необходимо для корреляции Пирсона
centered = user_item_matrix.sub(user_means, axis=0).fillna(0)
```

## 4. Вычисление схожести пользователей (корреляция Пирсона)

```python
from sklearn.metrics.pairwise import pairwise_distances

# Матрица расстояний (1 - корреляция)
distance_matrix = pairwise_distances(centered, metric='correlation')
# Превращаем в схожесть
user_sim_matrix = 1 - distance_matrix
# Диагональ = 1, отрицательные значения обнуляем
np.fill_diagonal(user_sim_matrix, 1.0)
user_sim_matrix[user_sim_matrix < 0] = 0

# Преобразуем в DataFrame для удобного доступа
user_sim_df = pd.DataFrame(
    user_sim_matrix,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)
print("Матрица схожести пользователей:", user_sim_df.shape)
```

## 5. Функция рекомендаций

```python
def get_recommendations(user_id, train, user_sim_df, user_item_matrix, k=10, n_neighbors=50):
    """
    Возвращает список из k рекомендованных фильмов для пользователя.
    - user_id: целевой пользователь
    - train: не используется, но оставлен для совместимости
    - user_sim_df: DataFrame схожести пользователей
    - user_item_matrix: матрица оценок (пользователи × фильмы)
    - k: количество рекомендаций
    - n_neighbors: сколько похожих пользователей учитывать
    """
    # Если пользователь новый – возвращаем популярные фильмы
    if user_id not in user_sim_df.index:
        popular = user_item_matrix.mean(axis=0).sort_values(ascending=False).head(k).index
        return list(popular)
    
    # Фильмы, которые пользователь уже оценил
    seen_movies = user_item_matrix.loc[user_id].dropna().index
    
    # Получаем топ n_neighbors наиболее похожих пользователей (исключая себя)
    sim_series = user_sim_df.loc[user_id].drop(user_id)
    sim_series = sim_series[sim_series > 0].sort_values(ascending=False).head(n_neighbors)
    
    if len(sim_series) == 0:
        # Нет похожих – возвращаем популярные
        popular = user_item_matrix.mean(axis=0).sort_values(ascending=False).head(k).index
        return list(popular)
    
    # Оценки похожих пользователей для всех фильмов
    neighbors_ratings = user_item_matrix.loc[sim_series.index]
    
    # Взвешенная сумма (веса = схожесть)
    weighted_sum = neighbors_ratings.T.dot(sim_series)
    total_weight = sim_series.sum()
    pred_scores = weighted_sum / total_weight
    
    # Убираем уже просмотренные фильмы
    pred_scores = pred_scores.drop(index=seen_movies, errors='ignore')
    
    # Сортируем и берём топ-k
    recommendations = pred_scores.sort_values(ascending=False).head(k).index.tolist()
    return recommendations
```

## 6. Оценка качества

```python
metrics = Metrics()
total_prec = 0
total_rec = 0
total_ndcg = 0

# Берём первых 1000 пользователей (как в примере content‑based)
users_sample = users['userId'].unique()[:1000]

for user in users_sample:
    recommended = get_recommendations(
        user, train, user_sim_df, user_item_matrix, k=10, n_neighbors=50
    )
    
    relevant = test[
        (test.userId == user) & (test.rating >= 4)
    ]['movieId'].tolist()
    
    if not len(relevant):
        continue
    
    precision, recall, ndcg = metrics.get_all(recommended, relevant, 10)
    total_prec += precision
    total_rec += recall
    total_ndcg += ndcg

# Вывод результатов
print(f"total precision: {total_prec / 1000}")
print(f"total recall: {total_rec / 1000}")
print(f"total ndcg: {total_ndcg / 1000}")
```
